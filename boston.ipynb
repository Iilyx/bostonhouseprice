Зависимые и независимые функции

```python
X=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
X_test
```

 Стандартизируем их

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

## Обучение

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
regression.coef_,regression.intercept_
```

```python
# параметры ар каким обучена
```

```python
regression.get_params()
```

```python
reg_pred=regression.predict(X_test)
```

```python
reg_pred
```

```python
plt.scatter(y_test,reg_pred) # линейный график хорошо
```

```python
residuals=y_test-reg_pred #остатки
```

```python
residuals
```

```python
sns.displot(residuals,kind="kde")
```

```python
## предсказания и остаток
```

```python
plt.scatter(reg_pred,residuals)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
print(mean_absolute_error(y_test,reg_pred),mean_squared_error(y_test,reg_pred),np.sqrt(mean_squared_error(y_test,reg_pred)),sep="\n")
```

R и R^2

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(y_test,reg_pred)
score
```

скорректированный r2

```python
1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
